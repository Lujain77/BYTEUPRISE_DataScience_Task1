```python
import pandas as pd
# The dataset
file_path = '/content/svm_data.csv'
data = pd.read_csv(file_path)

data.head()
```

```python
import matplotlib.pyplot as plt
# histogram
plt.figure(figsize=(10, 5))
plt.hist(data['x1'], bins=20, edgecolor='black',color='pink')
plt.title('Distribution of x1')
plt.xlabel('x1')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(data['x2'], bins=20, edgecolor='black',color='pink')
plt.title('Distribution of x2')
plt.xlabel('x1')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
# bar chart
plt.figure(figsize=(10, 5))
data['y'].value_counts().plot(kind='bar', color=['blue', 'purple'])
plt.title('Counts of y Categories')
plt.xlabel('y')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```
